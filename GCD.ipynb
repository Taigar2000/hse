```python
import numpy as np
import matplotlib.pyplot as plt
import copy
from sympy import *

fl=open("result.txt","w")


def sum(a):
    b=0
    for i in a:
        b+=i
    return b

def cut(f, counter=-1):
    if(len(f)==0): return []
    i=0
    while(f[i]==0):
        i+=1
        if(i==counter): break
        if(i==len(f)): return []
    return f[i::]

def analit(a,st=0):
    b=len(a)-1+st
    s=""
    for i in a:
        if(i!=0):
            if(i!=1 or b==0 or 1):
                s+=str(i)
            if(b>0):s+="*x**"+str(b)
            s+="+"
        b-=1
    if(len(s)==0):s="0+"
    return str(simplify(s[0:-1:]))


def GCD(f,g,mod,show=1,write=1):
    #if(len(f)<len(g)):f,g=g,f
    if(show):print("Counting for mod "+str(mod))
    if(write):fl.write("Counting for mod "+str(mod))
    retarr=[]
    retarr.append([copy.deepcopy(f),[-0],1,0])
    if(show):print(str(retarr[-1]))
    if(write):fl.write(str(retarr[-1]))
    retarr.append([copy.deepcopy(g),[-0],0,1])
    if(show):print(str(retarr[-1]))
    if(write):fl.write(str(retarr[-1]))


    g=cut(g)
    while(len(g)>0):
        n = len(g)#min(len(f),len(g))
        s=""
        ka=[]
        uv=[]
        while(len(f)>=len(g)):
            k=0
            iter=0
            
            if(show):
                print("\n\t"+analit(f))
            if(write):
                fl.write("\n\t"+analit(f))
                
            while(f[0]>0):
                k+=1
                f[0]=(f[0]-g[0]+mod)%mod
                iter+=1
                if(iter==mod+1):
                    break
                    #retarr.append([copy.deepcopy(f),"nan","nan","nan"])
                    #return retarr
            psa=[(g[i]*k)%mod for i in range (n)]
            for i in range(1,n):
                f[i]=(f[i]-g[i]*k+mod)%mod
            
            s+=str(k)
            ka.append(k)
            d=len(f)-n

            f=cut(f,1)        
            
            if(show):
                print("\n\t-"+analit(psa,d))
                print("\n\t="+analit(f))
            if(write):
                fl.write("\n\t-"+analit(psa,d))
                fl.write("\n\t="+analit(f)+"\n")
                
            
            if(d>0):s+="x**"+str(d)
            s+="+"
        s=s[0:-1:]
        if(len(ka)==0):ka=[0]
        f=cut(f)
        #retarr.append([copy.deepcopy(f),s,"("+str(retarr[-2][2])+")-("+s+")*("+str(retarr[-1][2])+")","("+str(retarr[-2][3])+")-("+s+")*("+str(retarr[-1][3])+")"])
        retarr.append([copy.deepcopy(f),ka,simplify("("+str(retarr[-2][2])+")-("+analit(ka)+")*("+str(retarr[-1][2])+")"),simplify("("+str(retarr[-2][3])+")-("+analit(ka)+")*("+str(retarr[-1][3])+")")])
        if(show):print(str(retarr[-1]))
        if(write):fl.write(str(retarr[-1])+"\n")
    #     print(str(f))
    #     fl.write(str(f))
    #     print(g)
    #     fl.write(str(g))
    #     print(k)
    #     fl.write(str(k))
    #     print((g*k)%mod)
    #     fl.write(str((g*k)%mod))

        fl.flush()

        f,g=g,f
    
    if(show):print("End of counting")
    if(write):fl.write("End of counting"+"\n")
    return retarr

def rev(f,mod,show,write):
    g=[mod for i in f]
    d=GCD(g,f,mod,show,write)
    flagf=1
    for i in d:
        if(flagf):
            flagf=0
            continue
        if(len(i[0])==1):
            b=copy.deepcopy(i)
            print (b)
            sc=1
            for j in range (1,mod+1):
                sc=j
                if(b[0][0]==1): 
                    return "("+str(b[-1])+"*"+str(sc)+")"
                    break
                b[0][0]=(b[0][0]+i[0][0])%mod
                print(b)
    return "nan"

if __name__=="__main__":

    #f=np.array([4,4,4,2,2,0,3,4,1])
    #g=np.array([3,4,0,2,1,2])
    f=np.array([4,2,1,2,2,4,4,1,4])
    g=np.array([3,2,4,0,3,1])
    mod=5#np.array([3,3,3,3,3])

    x=symbols("x")
    
    res=GCD(copy.deepcopy(f),copy.deepcopy(g),mod)
    print(simplify("("+str(simplify("("+str(res[-2][-2])+")*("+str(analit(res[0][0]))+") + ("+str(res[-2][-1])+")*("+str(analit(res[1][0]))+")"))+")%"+str(mod)))
    fl.write("("+str(res[-1][-2])+")*("+str(analit(res[0][0]))+") + ("+str(res[-1][-1])+")*("+str(analit(res[1][0]))+")")
    #Obratnii k f
    print("Obratnii k f = "+str(rev(f,mod,0,0)))
    fl.write("Obratnii k f = "+str(rev(f,mod,0,0)))
    #Obratnii k g
    print("Obratnii k g = "+str(rev([3,2],9,1,0)))
    fl.write("Obratnii k g = "+str(rev(g,mod,0,0)))


fl.close()
```
