```python
import numpy as np
import matplotlib.pyplot as plt
import copy
```

```python
#Funcs

def printf(fl,s):
    print(s,end="")
    try: fl.write(s)
    except: pass

def matrtostr(a, d=1, ri=0):
    s="$\\begin{pmatrix} \n"
    for i in a:
        for k in i:
            j=k
            if(ri): j=int(round(j))
            if(d==1): s+=str(j)+" & "
            else: s+="\\frac{"+str(j)+"}{"+str(d)+"}"+" & "
        s=s[:-2:]+"\\\\ \n"
    s+="\\end{pmatrix}$ \n"
    return s

def matrtostrds(a, astr, d=1, ri=0):
    s="$\\begin{pmatrix} \n"
    for i in range(len(a)):
        for k in range(len(a[i])):
            j=a[i][k]
            if(ri): j=int(round(j))
            if(d==1): s+=str(j)
            else: s+="\\frac{"+str(j)+"}{"+str(d)+"}"
            try:s+=astr[i][k]
            except: pass
            s+=" & "
        s=s[:-2:]+"\\\\ \n"
    s+="\\end{pmatrix}$ \n"
    return s

def roundm(ai):
    aj=[]
    for i in range(len(ai)):
        aj.append([])
        for j in range(len(ai[i])):
            aj[i].append(int(round(ai[i][j])))
            #print(aj[i][j])
    return aj
```

```python
#Init 

fl=open("GER.txt","w")

a=np.array([[-5,-2,-1,-5],
            [ 1, 0,-4,-5],
            [-2,-3, 1, 5],
            [-3, 2, 2,-4]])

b=np.array([[-4,-5, 0,-8],
            [ 4,-3,-3,11],
            [-2,-3, 2,-6],
            [-5,-4,-4,-6]])

# printf(fl,matrtostr(a))
# printf(fl,matrtostr(b))
```

```python
# N1

text1="""N1. Пусть $Q$ - искомая матрица в базисе, в котором даны координаты векторов в условии. \\
    Пусть матрица $A$ составлена из столбцов координат векторов $a_1, a_2, a_3, a_4$ соответственно. \\
    Пусть матрица $B$ составлена из столбцов координат векторов $b_1, b_2, b_3, b_4$ соответственно. \\
    
    \\begin{equation*}
        \\begin{cases}
            Qa_1=b_1
            Qa_2=b_2
            Qa_3=b_3
            Qa_4=b_4
        \\end{cases}
    \\end{equation*}
    
    Тогда: $Q \\cdot A = B$ \\\\\n
    """
printf(fl,text1)
printf(fl,"$A=$")
printf(fl,matrtostr(a)+"\n")
printf(fl,"$B=$")
printf(fl,matrtostr(b)+"\n")
deta=int(round(np.linalg.det(a)))
printf(fl,"$ det A = "+str(deta)+" \\neq 0 $\\\\\n")
printf(fl,"$A^{-1}$ определена,т.е.:\\\\\n")
ai=roundm(np.linalg.inv(a)*deta)
q=np.dot(b,ai)
printf(fl,"$Q=B \cdot A^{-1}=$"+matrtostr(b)+" $\\cdot$ "+matrtostr(a)[:-3]+"^{-1} =$ " + matrtostr(b) + "$ \\cdot (\\frac{1}{79}) \\cdot$" + matrtostr(ai) + "$ = (\\frac{1}{79}) \\cdot$" + matrtostr(q)+"\\\\\n")

printf(fl,"Найдём размерность ядра и размерность образа: \\\\\n")
printf(fl,"Пусть мы действуем в линейном пространстве V, следовательно $dim V \\leq 4$. \\\\\n")
printf(fl,"Так как $(a_1, a_2, a_3, a_4)$ линейно независимые (т.к. $det A \\neq 0$), то $dim V = 4$. \\\\\n")
printf(fl,"Так как $dim ImQ + dim Ker Q = dim V$, значит: \\\\\n")
printf(fl,"$dim Im Q + dim Ker Q = 4$. \\\\\n")
printf(fl,"Пусть $x \\in Ker Q$. Тогда $Q \\cdot x =$ "+matrtostr([[0],[0],[0],[0]])+"\\\\\n")
printf(fl,"Координаты $\\forall x \in Ker Q $ удовлетворяют данной СЛАУ и любое частное решение СЛАУ задаёт вектор $ \in Ker Q $. \\\\\n")
printf(fl,"Определим ранг СЛАУ (он будет равен рангу матрицы $Q$):")
printf(fl,"Так как $det$"+matrtostr(q)+" $=$ " + str(int(round(np.linalg.det(q))))+"")
printf(fl,", то $Rg Q < 4$ \\\\\n")
q3=np.array(q[0:3,0:3])
#print(q[0:3,0:3])
printf(fl,"Так как $det$"+matrtostr(q3)+" = " + str(int(round(np.linalg.det(q3))))+"")
printf(fl,", то $Rg Q \\geq 3$")
printf(fl,"Значит ранга СЛАУ $= 3$, и СЛАУ содержит только одну свободную переменную, а следовательно ФСР состоит из 1-го вектора. \\\\\n")
printf(fl,"Следовательно, $dim$ пространства решений СЛАУ $ = 1$, и $dim Ker Q = 1$. \\\\\n")
printf(fl,"$dim Im Q = dim V - dim Ker Q = 4 - 1 = 3$ \\\\\n")
printf(fl,"Ответ: \\\\\n")
printf(fl,matrtostr(q,d=deta)+"\\\\\n")
printf(fl,"Размерность ядра = 1 \\\\\n")
printf(fl,"Размерность образа = 1 \\\\\n")
```
