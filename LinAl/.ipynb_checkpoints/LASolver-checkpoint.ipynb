```python
import numpy as np
import matplotlib.pyplot as plt
import copy
from sympy import *
```

```python
#Funcs

def printf(fl,s):
    print(s,end="")
    try: fl.write(s)
    except: pass

def matrtostr(a, d=1, ri=0):
    s="$\\begin{pmatrix} \n"
    for i in a:
        for k in i:
            j=k
            if(ri): j=int(round(j))
            if(d==1): s+=str(j)+" & "
            else: s+="\\dfrac{"+str(j)+"}{"+str(d)+"}"+" & "
        s=s[:-2:]+"\\\\ \n"
    s+="\\end{pmatrix}$ \n"
    return s

def matrtostrds(a, astr, d=1, ri=0):
    s="\\begin{vmatrix} \n"
    for i in range(len(a)):
        for k in range(len(a[i])):
            j=a[i][k]
            s2=""
            if(ri): j=int(round(j))
            if(d==1): s2+=str(j)
            else: s2+="\\dfrac{"+str(j)+"}{"+str(d)+"}"
            try:s2+=astr[i][k]
            except: pass
            s+=s2.rjust(15,' ')+" & "
        s=s[:-2:]+"\\\\ \n"
    s+="\\end{vmatrix} \n"
    return s

def roundm(ai):
    aj=[]
    for i in range(len(ai)):
        aj.append([])
        for j in range(len(ai[i])):
            aj[i].append(int(round(ai[i][j])))
            #print(aj[i][j])
    return aj

def srez(a,ai,aj):
    ar=[]
    it=0
    ai.sort()
    aj.sort()
    
    iti=0
    for i in range(len(a)):
        if(iti<len(ai) and i==ai[iti]): 
            iti+=1
            continue
        ar.append([])
        itj=0
        for j in range(len(a[i])):
            if(itj<len(aj) and j==aj[itj]): 
                itj+=1
                continue
            ar[it].append(a[i][j])
        it+=1
    return ar
```

```python
#Init 

fl=open("GER.txt","w")

a=np.array([[-5,-2,-1,-5],
            [ 1, 0,-4,-5],
            [-2,-3, 1, 5],
            [-3, 2, 2,-4]])

b=np.array([[-4,-5, 0,-8],
            [ 4,-3,-3,11],
            [-2,-3, 2,-6],
            [-5,-4,-4,-6]])

# printf(fl,matrtostr(a))
# printf(fl,matrtostr(b))
```

```python
# N1

text1="""N1. Пусть $Q$ - искомая матрица в базисе, в котором даны координаты векторов в условии. \\
    Пусть матрица $A$ составлена из столбцов координат векторов $a_1, a_2, a_3, a_4$ соответственно. \\
    Пусть матрица $B$ составлена из столбцов координат векторов $b_1, b_2, b_3, b_4$ соответственно. \\
    
    \\begin{equation*}
        \\begin{cases}
            Qa_1=b_1
            Qa_2=b_2
            Qa_3=b_3
            Qa_4=b_4
        \\end{cases}
    \\end{equation*}
    
    Тогда: $Q \\cdot A = B$ \\\\\n
    """
printf(fl,text1)
printf(fl,"$A=$")
printf(fl,matrtostr(a)+"\n")
printf(fl,"$B=$")
printf(fl,matrtostr(b)+"\n")
deta=int(round(np.linalg.det(a)))
printf(fl,"$ det A = "+str(deta)+" \\neq 0 $\\\\\n")
printf(fl,"$A^{-1}$ определена,т.е.:\\\\\n")
ai=roundm(np.linalg.inv(a)*deta)
q=np.dot(b,ai)
printf(fl,"$Q=B \cdot A^{-1}=$"+matrtostr(b)+" $\\cdot$ "+matrtostr(a)[:-3]+"^{-1} =$ " + matrtostr(b) + "$ \\cdot (\\dfrac{1}{79}) \\cdot$" + matrtostr(ai) + "$ = (\\dfrac{1}{79}) \\cdot$" + matrtostr(q)+"\\\\\n")

printf(fl,"Найдём размерность ядра и размерность образа: \\\\\n")
printf(fl,"Пусть мы действуем в линейном пространстве V, следовательно $dim V \\leq 4$. \\\\\n")
printf(fl,"Так как $(a_1, a_2, a_3, a_4)$ линейно независимые (т.к. $det A \\neq 0$), то $dim V = 4$. \\\\\n")
printf(fl,"Так как $dim ImQ + dim Ker Q = dim V$, значит: \\\\\n")
printf(fl,"$dim Im Q + dim Ker Q = 4$. \\\\\n")
printf(fl,"Пусть $x \\in Ker Q$. Тогда $Q \\cdot x =$ "+matrtostr([[0],[0],[0],[0]])+"\\\\\n")
printf(fl,"Координаты $\\forall x \in Ker Q $ удовлетворяют данной СЛАУ и любое частное решение СЛАУ задаёт вектор $ \in Ker Q $. \\\\\n")
printf(fl,"Определим ранг СЛАУ (он будет равен рангу матрицы $Q$):")
printf(fl,"Так как $det$"+matrtostr(q)+" $=$ " + str(int(round(np.linalg.det(q))))+"")
printf(fl,", то $Rg Q < 4$ \\\\\n")
q3=np.array(q[0:3,0:3])
#print(q[0:3,0:3])
printf(fl,"Так как $det$"+matrtostr(q3)+" = " + str(int(round(np.linalg.det(q3))))+"")
printf(fl,", то $Rg Q \\geq 3$")
printf(fl,"Значит ранга СЛАУ $= 3$, и СЛАУ содержит только одну свободную переменную, а следовательно ФСР состоит из 1-го вектора. \\\\\n")
printf(fl,"Следовательно, $dim$ пространства решений СЛАУ $ = 1$, и $dim Ker Q = 1$. \\\\\n")
printf(fl,"$dim Im Q = dim V - dim Ker Q = 4 - 1 = 3$ \\\\\n")
printf(fl,"Ответ: \\\\\n")
printf(fl,matrtostr(q,d=deta)+"\\\\\n")
printf(fl,"Размерность ядра = 1 \\\\\n")
printf(fl,"Размерность образа = 1 \\\\\n")
```

```python
# N2
printf(fl,"\n\n\nN2 Найдём собственные значения \\\\\n")
le=np.array([["-\\lambda","","",""],
            ["","-\\lambda","",""],
            ["","","-\\lambda",""],
            ["","","","-\\lambda"]])
a=np.array([[   9,  1, 0,0],
            [ -43, -6, 1,0],
            [-125,-32, 9,0],
            [-205,-55,10,4]])
printf(fl,"$\\chi_A (\\lambda) = det(A-\\lambda E) = ")
printf(fl,matrtostrds(a,le))

ai=[3]
aj=[3]
a012012=np.array(srez(a,ai,aj))
le012012=np.array(srez(le,ai,aj))

printf(fl," = ")
printf(fl,"(4-\\lambda) \\cdot ("+matrtostrds(a012012,le012012))

#return



ai=[0]
aj=[0]
a1212=np.array(srez(a012012,ai,aj))
le1212=np.array(srez(le012012,ai,aj))


ai=[0]
aj=[1]
a1202=np.array(srez(a012012,ai,aj))
le1202=np.array(srez(le012012,ai,aj))

printf(fl," = ")
printf(fl,"(4-\\lambda) \\cdot \left( (9-\\lambda) \\cdot "+matrtostrds(a1212,le1212))
printf(fl," - "+matrtostrds(a1202,le1202) + "\right) = ")
stro="(4-\\lambda) \\cdot ( (9-\\lambda) \\cdot ( (-6-\\lambda) \\cdot (9-\\lambda) + 32 ) + 43 \\cdot (9-\\lambda) - 125 )"
printf(fl,stro+" = ")
strp=stro.replace("\\lambda","x").replace("\\cdot","*")
#print("\n\n"+strp)
x=symbols("x")
strp=str(factor(strp))
stro=strp.replace("x","\\lambda").replace("**","^").replace("*","\\cdot")
printf(fl,stro + "$ \\\\\n")

text2="""Так как $\\chi_A (\\lambda) =(\\lambda - 4)^4 $, то имеется всего одно собственное значение $ = 4$ с алгебраической кратностью 4. \\\\

Определим число жордановых клесок, соответствующих собственному значению. \\\\
Все жордановы клетки отвечают собственному значению 4 (так как алгебраическоя кратность равна 4), и могут иметь порядок 1, 2, 3, 4. \\\\
"""
printf(fl,text2)
a4=[]
for i in range(len(a)):
    a4.append([])
    for j in range(len(a[i])):
        a4[i].append(a[i][j])
        if(i==j):a4[i][j]-=4
            
printf(fl,"$A-4E = $"+matrtostr(a4)+"\\\\\n")
r0=4
r1=np.linalg.matrix_rank(a4)
r2=np.linalg.matrix_rank(np.dot(a4,a4))
r3=np.linalg.matrix_rank(np.dot(a4,np.dot(a4,a4)))
r4=np.linalg.matrix_rank(np.dot(a4,np.dot(a4,np.dot(a4,a4))))
r5=np.linalg.matrix_rank(np.dot(a4,np.dot(a4,np.dot(a4,np.dot(a4,a4)))))
#print(r0,r1,r2,r3)
#printf(fl,"$(A-4E) = $"+matrtostr(np.dot(a4,a4))+"\\\\\n")
printf(fl,"$r^k = Rg(A-4E)^k$ и $r^0 = 4$ \\\\\n")
printf(fl,"Число клеток 1x1 = $r^0 - 2r^1 + r^2 = "+str(r0)+" - "+str(2*r1)+" + "+str(r2)+" = "+str(r0-2*r1+r2)+"$\\\\\n")
printf(fl,"Число клеток 2x2 = $r^1 - 2r^2 + r^3 = "+str(r1)+" - "+str(2*r2)+" + "+str(r3)+" = "+str(r1-2*r2+r3)+"$\\\\\n")
printf(fl,"Число клеток 3x3 = $r^2 - 2r^3 + r^4 = "+str(r2)+" - "+str(2*r3)+" + "+str(r4)+" = "+str(r2-2*r3+r4)+"$\\\\\n")
printf(fl,"Число клеток 4x4 = $r^3 - 2r^4 + r^5 = "+str(r3)+" - "+str(2*r4)+" + "+str(r5)+" = "+str(r3-2*r4+r5)+"$\\\\\n")
printf(fl,"Cоставим искомую ЖНФ из жордановых клеток 1х1 и 3х3. \\\\\n")
retm=np.array([[4,0,0,0],
               [0,4,1,0],
               [0,0,4,1],
               [0,0,0,4]])
printf(fl,"Ответ:  "+matrtostr(retm))
```
