```python
import numpy as np
import matplotlib.pyplot as plt
import copy
from sympy import *

fl=open("result.txt","w")


def sum(a):
    b=0
    for i in a:
        b+=i
    return b

def cut(f, counter=-1):
    if(len(f)==0): return []
    i=0
    while(f[i]==0):
        i+=1
        if(i==counter): break
        if(i==len(f)): return []
    return f[i::]

def analit(a,st=0):
    b=len(a)-1+st
    s=""
    for i in a:
        if(i!=0):
            if(i!=1 or b==0 or 1):
                s+=str(i)
            if(b>0):s+="*x**"+str(b)
            s+="+"
        b-=1
    if(len(s)==0):s="0+"
    return str(simplify(s[0:-1:]))


def GCD(f,g,mod,show=1,write=1):
    #if(len(f)<len(g)):f,g=g,f
    if(show):print("Counting for mod "+str(mod))
    if(write):fl.write("\n\nCounting for mod "+str(mod)+"\n")
    retarr=[]
    retarr.append([copy.deepcopy(f),[-0],1,0])
    if(show):print(str(retarr[-1]))
    if(write):fl.write(str(retarr[-1])+"\n")
    retarr.append([copy.deepcopy(g),[-0],0,1])
    if(show):print(str(retarr[-1]))
    if(write):fl.write(str(retarr[-1])+"\n")


    g=cut(g)
    while(len(g)>0):
        n = len(g)#min(len(f),len(g))
        s=""
        ka=[]
        uv=[]
        breakf=0
        while(len(f)>=len(g)):
            k=0
            iter=0
            
            if(show):
                print("\n\t"+analit(f))
            if(write):
                fl.write("\n\t"+analit(f))
                
            while(f[0]>0):
                k+=1
                f[0]=(f[0]-g[0]+mod)%mod
                iter+=1
                if(iter==mod):
                    breakf=1
                    k=1
                    break
                    #retarr.append([copy.deepcopy(f),"nan","nan","nan"])
                    #return retarr
            
            psa=[(g[i]*k)%mod for i in range (n)]
            for i in range(1,n):
                f[i]=(f[i]-g[i]*k+mod)%mod
#                 while(k<0):
#                     k=(k+mod)%mod
            s+=str(k)
            ka.append(k)
            d=len(f)-n

            f=cut(f,1)        
            
            if(show):
                print("\n\t-"+analit(psa,d))
                print("\n\t="+analit(f))
            if(write):
                fl.write("\n\t-"+analit(psa,d))
                fl.write("\n\t="+analit(f)+"\n")
                
            
            if(d>0):s+="x**"+str(d)
            s+="+"
            if(breakf):break
        s=s[0:-1:]
        if(len(ka)==0):ka=[0]
        f=cut(f)
        #retarr.append([copy.deepcopy(f),s,"("+str(retarr[-2][2])+")-("+s+")*("+str(retarr[-1][2])+")","("+str(retarr[-2][3])+")-("+s+")*("+str(retarr[-1][3])+")"])
        retarr.append([copy.deepcopy(f),ka,simplify("("+str(retarr[-2][2])+")-("+analit(ka)+")*("+str(retarr[-1][2])+")"),simplify("("+str(retarr[-2][3])+")-("+analit(ka)+")*("+str(retarr[-1][3])+")")])
        if(show):print(str(retarr[-1]))
        if(write):fl.write(str(retarr[-1])+"\n")
    #     print(str(f))
    #     fl.write(str(f))
    #     print(g)
    #     fl.write(str(g))
    #     print(k)
    #     fl.write(str(k))
    #     print((g*k)%mod)
    #     fl.write(str((g*k)%mod))

        if(write): fl.flush()

        if(breakf and len(f)>=len(g)):break
        
        f,g=g,f
        
        
    if(show):print("End of counting")
    if(write):fl.write("End of counting"+"\n")
    return retarr

def rev(f,g,mod,show,write):
    #g=[mod for i in f]
    f=[f]
    g=[g]
    d=GCD(g,f,mod,show,write)
    flagf=1
    for i in ((d[::-1])[::]):
        try: 
#         if(flagf):
#             flagf=0
#             continue
         if(len(i[0])==1):           
            b=copy.deepcopy(i)
            #print (b)
            sc=1
            for j in range (1,mod+1):
                sc=j
                if(b[0][-1]==1): 
                    return int(b[-1])*sc
                    break
                b[0][-1]=(b[0][-1]+i[0][-1])%mod
                #print(b)
        except:
            pass
    return "nan"



if (0 and __name__=="__main__"):

    #f=np.array([4,4,4,2,2,0,3,4,1])
    #g=np.array([3,4,0,2,1,2])
    p=np.array([4,2,1,2,2,4,4,1,4]) #P(x)
    f=np.array([3,2,4,0,3,1]) #F(x)
    mod=5#np.array([3,3,3,3,3])

    x=symbols("x")
    
    res=GCD(copy.deepcopy(p),copy.deepcopy(f),mod)
    print("GCD(P,F): "+str(simplify("("+str(simplify("("+str(res[-2][-2])+")*("+str(analit(res[0][0]))+") + ("+str(res[-2][-1])+")*("+str(analit(res[1][0]))+")"))+")%"+str(mod))))
    fl.write("GCD(P,F): "+str(simplify("("+str(simplify("("+str(res[-2][-2])+")*("+str(analit(res[0][0]))+") + ("+str(res[-2][-1])+")*("+str(analit(res[1][0]))+")"))+")%"+str(mod))))
    #Obratnii k f
    print("Obratnii k P = "+str(rev(copy.deepcopy(p),mod,1,1)))
    fl.write("Obratnii k P = "+str(rev(copy.deepcopy(p),mod,0,0)))
    #Obratnii k g
    print("Obratnii k F = "+str(rev(copy.deepcopy(f),mod,1,1)))
    fl.write("Obratnii k F = "+str(rev(copy.deepcopy(f),mod,0,0)))

    #print("Obratnii k g = "+str(rev([3,2],9,1,0)))

fl.close()
```

```python
def add(p,q=[],a=0, show=0,write=0,fl=0,mod=1):
    s=""
    s2=""
    #print(p,q)
    if(q!=[] and (not(p[0]==q[0] and p[1]==q[1]))):
        #print(1.1)
        l=(int(q[1]-p[1]))*rev(int(q[0]-p[0]),mod,mod,0,0)
        #print(1.2)
        s="((("+str(q[1])+")-("+str(p[1])+"))/(("+str(q[0])+")-("+str(p[0])+")))"
        s2="((("+str(q[1])+")-("+str(p[1])+"))*("+str(rev(int(q[0]-p[0]),mod,mod,0,0))+"))"
#        if(show): print("("+str(q[1])+"-"+str(p[1])+")/("+str(q[0])+"-"+str(p[0])+")")
#        if(write): fl.write("("+str(q[1])+"-"+str(p[1])+")/("+str(q[0])+"-"+str(p[0])+")\n")
    #print(2)
    if(q==[] or (p[0]==q[0] and p[1]==q[1])):
        l=(int(3*(p[0]**2)+a))*rev(int(2*p[1]),mod,mod,0,0)
        s="((3*("+str(p[0])+")**2+("+str(a)+"))/(2*("+str(p[1])+")))"
        s2="((3*("+str(p[0])+")**2+("+str(a)+"))*("+str(rev(int(2*p[1]),mod,mod,0,0))+"))"
#        if(show): print("(3*"+str(p[0])+"**2+"+str(a)+")/(2*"+str(p[1])+")")
#        if(write): fl.write("(3*"+str(p[0])+"**2+"+str(a)+")/(2*"+str(p[1])+")")
    #print(3)
    xs=((int(l**2)-p[0]-q[0])%mod+mod)%mod
    so=s+"**2-("+str(p[0])+")-("+str(q[0])+")" + " =\n " + s2+"**2-("+str(p[0])+")-("+str(q[0])+")" + " =\n "+str(xs)
    if(show): print(so)
    if(write): fl.write(so)
    ys=((-p[1]+int(l*(p[0]-xs)))%mod+mod)%mod
    so="-("+str(p[1])+")+"+s+"*(("+str(p[0])+")-("+str(xs)+"))" + " =\n " + "-("+str(p[1])+")+"+s2+"*(("+str(p[0])+")-("+str(xs)+"))" + " =\n "+str(ys)
    if(show): print(so)
    if(write): fl.write(so)
    return [xs,ys]

def mult(p,a,n,show=0,write=0,fl=0,mod=1):
    ret=copy.deepcopy(p)
    flag=1
    for i in range(n):
        if(flag): flag=(flag^1)
        ret=add(ret,p,a,(show and i==n-1),(write and i==n-1),fl,mod)
    if(0 and flag):
        if(show): print("("+str(ret[0])+";",end="")
        if(write): fl.write("("+str(ret[0])+";")
        if(show): print(str(ret[1])+")")
        if(write): fl.write(str(ret[1])+")\n")
    return ret

def por(p,a,say=1,sayf=1,show=0,write=0,fl=0,mod=1,retno=1):
    ret=copy.deepcopy(p)
    flag=1
    i=1
    #try:
    if(1):
        while(1):
            #if(flag): flag=(show^1)
            if(say): print({i:ret[-1]})
            if(sayf): 
                fl.write(str({i:ret[-1]})+"\n")
                fl.flush()
            ret.append(add(ret[-1],p[0],a,show,write,fl,mod))
            if(retno and ret[-1][0]==p[0][0] and ret[-1][1]==p[0][1]): return {i:ret[-1]}
            i+=1
    #except:
    #    return i+1
    if(retno and ret[-1][0]==p[0][0] and ret[-1][1]==p[0][1]): return {i:ret[-1]}
    return "none"
```
